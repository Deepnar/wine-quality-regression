```python
from sklearn.model_selection import train_test_split

training, testing = train_test_split(wine,test_size=0.33,random_state=42)
```

```python
trainX=training.drop('quality',axis=1)
trainY=training['quality']
```

```python
testX=testing.drop('quality',axis=1)
testY=testing['quality']
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
trainX_scaled=scalar.fit_transform(trainX)
testX_scaled=scalar.transform(testX)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg_rfr=RandomForestRegressor()
reg_rfr.fit(trainX_scaled,trainY)
reg_rfr.predict(testX_scaled)
```

```python
print(reg_rfr.score(testX_scaled,testY))
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print("mean abseloute error ",  mean_absolute_error(testY,reg_rfr.predict(testX_scaled)))
print("mean squared error ", mean_squared_error(testY,reg_rfr.predict(testX_scaled)))
print("r2score ",  r2_score(testY,reg_rfr.predict(testX_scaled)))
```

```python
from sklearn.svm import SVR
reg_svr=SVR()
reg_svr.fit(trainX_scaled,trainY)
reg_svr.predict(testX_scaled)
```

```python
print("mean abseloute error ",  mean_absolute_error(testY,reg_svr.predict(testX_scaled)))
print("mean squared error ", mean_squared_error(testY,reg_svr.predict(testX_scaled)))
print("r2score ",  r2_score(testY,reg_svr.predict(testX_scaled)))
```

```python
from sklearn.model_selection import GridSearchCV

parameter={'kernel':['rbf'],'C':(0.1,1,10,100),'gamma':(1,0.1,0.01,0.001)}



clf = GridSearchCV(reg_svr,parameter,cv=5)
clf.fit(trainX_scaled, trainY)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
reg_gbr=GradientBoostingRegressor()
reg_gbr.fit(trainX_scaled,trainY)
reg_gbr.predict(testX_scaled)
```

```python
print("mean abseloute error ",  mean_absolute_error(testY,reg_gbr.predict(testX_scaled)))
print("mean squared error ", mean_squared_error(testY,reg_gbr.predict(testX_scaled)))
print("r2score ",  r2_score(testY,reg_gbr.predict(testX_scaled)))
```
